```python
#Dropping columns which are not used for analysis
df = df.drop('State Code', axis=1)
df = df.drop('Date of status changed ', axis=1)
df = df.drop('Backup Notes', axis=1)
```

```python
#Now checking for number of null value columns
df.isnull().sum(axis=0)
```

```python
#Dropping rows having more than 7 null values 
df=df[df.isnull().sum(axis=1)<8]
```

Dimension of dataset left is 4365 * 12

```python
# Again finding percentage of missing value in each column
round(100*(df.isnull().sum()/len(df.index)), 2) 
```

```python
df['Age']=df['Age'].map({np.NaN:'0'}).fillna(df['Age'])
```

```python
df
```

As dataset is having larger percentage of null values, therefore we simply cannot remove it as it will lead to larger data loss. Therefore replacing null values with a common tag called "Not Known"

```python
df=df.replace(np.NaN,'Not Known')
```

```python
#Final check for null values in each column
df.isnull().sum(axis=0)
```

# Dataset is now cleaned and ready for analysis

```python
#Importing libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=df.replace('28-35','31')
```

```python
df['Age']=df['Age'].astype(str).astype(int)
```

```python
def age_buckets(x):
    if (x==0):
        return 'Not Known'
    elif (x<=18 and x>0):
        return '0-18'
    elif (x>18 and x<=45):
        return '19-45'
    elif (x>45 and x<=60):
        return '46-60'
    else:
        return '60+'
```

```python
df['Age']=df.Age.apply(age_buckets)
```

```python
#Age wise cases
df['Age'].value_counts()
```

```python
# Countplot of number of cases and age
sns.set(style="darkgrid")
plt.figure(figsize=(20, 8))
ax = sns.countplot(x=df['Age'],palette='Set1')
plt.title('Count of Covid-19 cases according to age')
plt.xlabel('Age Groups')
plt.ylabel('Number of cases')
plt.show()
```

Analysis :- People with age 19-45 are the one which are affected the most whereas people between 0-18 are affected least and age of major cases is not known.

```python
#State wise cases
df['Detected State'].value_counts()
```

```python
df['Detected State'].value_counts().plot(kind='barh', figsize=(8, 10),color="#48C9B0")
plt.xlabel("Count of covid-19 cases in each state", labelpad=14)
plt.ylabel("States of India", labelpad=18)
plt.title("Count of cases in each state", y=1);
```

Analysis:- From this plot it is analyzed that Maharashtra is having the maximum count of cases whereas states of north east are at lower risk.

```python
#Gender wise cases
df['Gender'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
count = df['Gender'].value_counts()
colors = ["#BB8FCE", "#58D68D", "#F5B041"]
lab = ['Not Known','Male','Female']
plt.pie(count,  colors=colors, labels=lab,autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Cases according to gender")
plt.show()
```

Analysis:- From this pie chart it is clearly visible that 23% Male are suffered whereas very less females are suffered and more than half affected people's gender is not known.

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20, 8))
sns.countplot(x="Age", hue="Gender" ,data=df,palette='Set2')
plt.title('Age,Gender versus number of cases')
plt.xlabel('Age')
plt.ylabel('Number of cases')
plt.show()
```

Analysis:- Major males affected are of age group 19-45

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10,20))
sns.countplot(y="Detected State", hue="Gender" ,data=df,palette='Set1')
plt.title('State,Gender versus income')
plt.xlabel('Number of cases')
plt.ylabel('State')
plt.show()
```

Analysis:- Males most affected are from Tamil Nadu state whereas females most affected are from Karnataka State.
